# EDA и эксперименты по прогнозированию просрочки задач

Ноутбук демонстрирует чтение синтетического датасета, базовый EDA и сравнение нескольких моделей для задачи бинарной классификации `is_overdue`.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

DATA_PATH = Path('../data/tasks_synthetic.csv')
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
print('Rows:', len(df))
print('Columns:', len(df.columns))
print('Overdue share:', round(df['is_overdue'].mean(), 4))
df.describe(include='all').T.head(20)
```

## Наблюдения по данным

- Класс умеренно несбалансирован: около 30% задач помечены как просроченные.
- Риск просрочки растёт у задач с блокерами, высоким приоритетом и низким прогрессом.
- Для сервиса особенно важны признаки состояния задачи в текущий момент, а не только её тип.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.countplot(data=df, x='priority', hue='is_overdue', ax=axes[0, 0])
axes[0, 0].set_title('Просрочка по приоритетам')
sns.boxplot(data=df, x='is_overdue', y='estimated_hours', ax=axes[0, 1])
axes[0, 1].set_title('Оценка трудоёмкости vs просрочка')
sns.boxplot(data=df, x='is_overdue', y='actual_progress', ax=axes[1, 0])
axes[1, 0].set_title('Фактический прогресс vs просрочка')
sns.barplot(data=df, x='has_blockers', y='is_overdue', ax=axes[1, 1])
axes[1, 1].set_title('Влияние блокеров')
plt.tight_layout()
```

```python
priority_risk = df.groupby('priority')['is_overdue'].mean().sort_values(ascending=False)
status_risk = df.groupby('status')['is_overdue'].mean().sort_values(ascending=False)
print('Risk by priority:')
print(priority_risk.round(3))
print()
print('Risk by status:')
print(status_risk.round(3))
```

## Эксперименты с моделями

Ниже зафиксированы метрики из обучающего скрипта `src/train.py`. Для финального сервиса выбрана модель по `F1`, так как в бизнес-сценарии важно не пропускать реально рискованные задачи.

```python
results = pd.DataFrame([
    {'model': 'LogisticRegression', 'accuracy': 0.6700, 'precision': 0.4637, 'recall': 0.6622, 'f1': 0.5455, 'roc_auc': 0.7270},
    {'model': 'RandomForest', 'accuracy': 0.7190, 'precision': 0.5421, 'recall': 0.3880, 'f1': 0.4522, 'roc_auc': 0.6986},
    {'model': 'GradientBoosting', 'accuracy': 0.7320, 'precision': 0.6148, 'recall': 0.2776, 'f1': 0.3825, 'roc_auc': 0.6946},
])
results.sort_values('f1', ascending=False)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=results.melt(id_vars='model', value_vars=['f1', 'roc_auc']), x='model', y='value', hue='variable')
plt.title('Сравнение моделей по F1 и ROC-AUC')
plt.xticks(rotation=15)
plt.tight_layout()
```

## Вывод

`LogisticRegression` выбрана как финальная модель, потому что она показала лучший `F1` и `Recall`. Это соответствует цели проекта: заранее выявлять потенциально просроченные задачи и использовать модель как систему раннего предупреждения.
