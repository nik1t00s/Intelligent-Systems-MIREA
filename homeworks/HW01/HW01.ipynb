## 1. Импорт библиотеки NumPy с общепринятым сокращением

```python
import numpy as np
```

## 2. Примеры создания массивов

```python
# Из обычных списков
list_data = [0, 1, 2, 3, 4]
array_from_list = np.array(list_data)
print(f"Исходный список: {list_data}")
print(f"Numpy массив: {array_from_list}")
print(f"Тип массива: {type(array_from_list)}\n")
```

```python
# С использованием функций `zeros`, `ones`, `arange` или аналогичных.
zeros_array = np.zeros(3)
ones_array = np.ones(5)
arange_array = np.arange(0, 8, 2)  # Массив от 0 до 8 с шагом 2

print("Массив из нулей:")
print(zeros_array)
print(f"Shape: {zeros_array.shape}, Dtype: {zeros_array.dtype}\n")

print("Массив из единиц:")
print(ones_array)
print(f"Shape: {ones_array.shape}, Dtype: {ones_array.dtype}\n")

print("Массив с помощью arange:")
print(arange_array)
print(f"Shape: {arange_array.shape}, Dtype: {arange_array.dtype}\n")
```

## 3. Демонстрация атрибутов

```python
# Вывод формы (shape) и типа данных (dtype) для одного из массивов.
print(f"Массив: {array_from_list}")
print(f"- Форма (shape): {array_from_list.shape}")
print(f"- Тип данных (dtype): {array_from_list.dtype}\n")
```

## 4. Примеры индексирования и срезов

```python
print(f"Исходный массив: {array_from_list}")

# Получение отдельного элемента
element = array_from_list[2]
print(f"Элемент на позиции [2]: {element}")

# Извлечение подмассива с помощью среза
subarray = array_from_list[1:4]
print(f"Срез [1:4]: {subarray}\n")
```

## 5. Примеры векторных операций и базовых статистик

```python
array_a = np.array([0, 1, 2, 3, 4])
array_b = np.array([0, 10, 20, 30, 40])

print(f"Массив A: {array_a}")
print(f"Массив B: {array_b}")

# Поэлементное сложение
sum_result = array_a + array_b
print(f"\nПоэлементное сложение (A + B): {sum_result}")

# Поэлементное умножение
product_result = array_a * array_b
print(f"Поэлементное умножение (A * B): {product_result}")
```

```python
# Вычисление средней величины (`mean`) или другой простой статистики.
print(f"Среднее значение (mean): {np.mean(array_a):.2f}")
```
